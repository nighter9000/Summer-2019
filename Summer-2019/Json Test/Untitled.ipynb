```python
import json
from collections import defaultdict
import numpy as np
import matplotlib.pyplot as plt

with open('Ethan.json') as json_file:
    data = json.load(json_file)
```

```python
#creates dictionary keys of unique each value in data with values of instances used
semisortedinfo = {}
res = defaultdict(list) 
for p in data:
    res = defaultdict(list) 
    for key, val in sorted(data[p].items()): 
        res[val].append(key)
    semisortedinfo.update({p:dict(res)})
```

```python
#unique values by field
print('Number of Unique Values in Each Field:\n')
for i in semisortedinfo:
    print('\t'+str(i)+'  '+str(len(semisortedinfo[i])))
```

```python
#unknowns
semisortedinfo['CMSPrimaryPrimaryDataset']['Unknown']
```

```python
lengths = []
for i in semisortedinfo[[*data][1]]:
    #print(i + '\t' + str(len(semisortedinfo[[*data][1]][i])))
    lengths.append(len(semisortedinfo[[*data][1]][i]))
    
y_pos = np.arange(len(semisortedinfo[[*data][1]])) 
# Create bars
plt.bar(y_pos, lengths)
# Create names on the x-axis
plt.xticks(y_pos, semisortedinfo[[*data][1]])
# Show graphic
plt.show()

```

```python
for p in data:
    lengths = []
    fig, ax = plt.subplots(figsize=(15, 15))
    for i in semisortedinfo[p]:
        #print(i + '\t' + str(len(semisortedinfo[[*data][1]][i])))
        lengths.append(len(semisortedinfo[p][i]))
    
    y_pos = np.arange(len(semisortedinfo[p])) 
    # Create bars
    plt.bar(y_pos, lengths)
    # Create names on the x-axis
    plt.xticks(y_pos, semisortedinfo[p])
    # Show graphic
    plt.title(p)
    plt.xticks(rotation = -90)
    plt.tight_layout()
    
    plt.show()
    
```
