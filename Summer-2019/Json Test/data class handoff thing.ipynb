```python
import json
from collections import defaultdict
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
with open('Ethan.json') as json_file:
    data = json.load(json_file)
```

```python
#creates dictionary keys of unique each value in data with values of instances used
semisortedinfo = {}
res = defaultdict(list) 
for p in data:
    res = defaultdict(list) 
    for key, val in sorted(data[p].items()): 
        res[val].append(key)
    semisortedinfo.update({p:dict(res)})
```

```python
#unique values by field
print('Number of Unique Values in Each Field:\n')
for i in semisortedinfo:
    print('\t'+str(i)+'  '+str(len(semisortedinfo[i])))
```

```python
for p in semisortedinfo:
    print('Number of Unique Values in '+str(p)+':\n')
    if (len(semisortedinfo[p])<50):
        for i in semisortedinfo[p]:
            print('\t'+str(i)+'  '+str(len(semisortedinfo[p][i])))
        print('')
```

```python
lengths = []
for i in semisortedinfo[[*data][1]]:
    #print(i + '\t' + str(len(semisortedinfo[[*data][1]][i])))
    lengths.append(len(semisortedinfo[[*data][1]][i]))
    
y_pos = np.arange(len(semisortedinfo[[*data][1]])) 
# Create bars
plt.bar(y_pos, lengths)
# Create names on the x-axis
plt.xticks(y_pos, semisortedinfo[[*data][1]])
# Show graphic
plt.show()

```

```python
for p in data:
    lengths = []
    fig, ax = plt.subplots(figsize=(15, 15))
    for i in semisortedinfo[p]:
        #print(i + '\t' + str(len(semisortedinfo[[*data][1]][i])))
        lengths.append(len(semisortedinfo[p][i]))
    
    y_pos = np.arange(len(semisortedinfo[p])) 
    # Create bars
    plt.bar(y_pos, lengths)
    # Create names on the x-axis
    plt.xticks(y_pos, semisortedinfo[p])
    # Show graphic
    plt.title(p)
    plt.xticks(rotation = -90)
    plt.tight_layout()
    
    #plt.show()
    
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tester = [['sd',3,5,'apple'],['la',3,5,'orange'],['sd',4,8,'apple'],['la',6,5,'orange'],['sd',2,5,'apple'],['sd',4,5,'apple'],['la',4,3,'orange'],['la',5,3,'orange']]

# generating a dataframe from the matrix 
test = pd.DataFrame(tester, columns = ['Site','weight','height','label'])

# defining predictors
X = test[['height','weight']]

# definig the target variable and mapping it to 1 for apple and 0 for orange
y = test['label'].replace({'apple':1, 'orange':0})

# instantiating the model
tree = DecisionTreeClassifier() 

# fitting the model
model = tree.fit(X,y)
```

```python
from sklearn.externals.six import StringIO  
from sklearn.tree import export_graphviz
import pydotplus
from IPython.display import Image


dot_data = StringIO()  

export_graphviz(
    model, 
    out_file = dot_data,  
    filled=True, rounded=True, proportion=False,
    special_characters=True, 
    feature_names=X.columns,
    class_names=["orange", "apple"]
)  

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  

Image(graph.create_png())
```

```python
#creates a dictionary to make each string into a number for the classifier
#each field will map one value to one number for example:
#{'T2_US_UCSD': 9} maps ucsd to 9 and is in the 'CMSSite' field

#the dictionary is used with the command df.replace(bigdict)
bigdict = {}

for p in data:
    if type(data[p]['1']) is not str:   
        continue
    if (len(semisortedinfo[p])/len(data[p])>.5):
        continue
    counter = 0
    littledict = {}
    for i in semisortedinfo[p]:
        littledict.update({i:counter})
        counter+=1
    bigdict.update({p:littledict})

#dictionary can be updated using the following method:    
#bigdict.update({'CMSPrimaryPrimaryDataset':{'tester':0}})
```

```python
target = 'RequestCpus'
predictors = []
for i in [*df.replace(bigdict)]:
    if i == target:
        continue
    predictors.append(i)

# defining predictors
X = df.replace(bigdict)[predictors]

# definig the target variable and mapping it to 1 for apple and 0 for orange
y = df.replace(bigdict)[target]

# instantiating the model
tree = DecisionTreeClassifier() 

# fitting the model
model = tree.fit(X,y)
```

```python

dot_data = StringIO()  

export_graphviz(
    model, 
    out_file = dot_data,  
    filled=True, rounded=True, proportion=False,
    special_characters=True, 
    feature_names=X.columns,
    class_names=["1", "4"]
)  

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  

Image(graph.create_png())
```
